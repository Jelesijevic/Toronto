```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import json
```

```python
postcodes_toronto= requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
#  Wikipedia page with List of Postcodes in Toronto
soup = BeautifulSoup(postcodes_toronto,'lxml')
```

```python
# Create three lists of data
postcodes_list=[]
borough_list=[]
neighborhood_list=[]
i=1
for tag in soup.table.find_all('td'):
    if i == 1:
        postcodes_list.append(tag.text)
    if i == 2:
        borough_list.append(tag.text)
    if i == 3: 
        neighborhood_list.append(tag.text)
    i = i+1
    if i==4:
        i=1
```

```python
# Creating the dataframe
df_pct = pd.DataFrame(columns=['Postcodes','Borough','Neighborhood'])
df_pct['Postcodes'] = postcodes_list
df_pct['Borough'] = borough_list
df_pct['Neighborhood'] = neighborhood_list
df_pct
```

```python
# Locating indexes'Not assigned' for Boroughs
indexes = df_pct[df_pct['Borough']=="Not assigned"].index
indexes
```

```python
# remove 'Not assigned' from Boroughs
df_pct.drop(df_pct.index[indexes], inplace=True)
```

```python
# replace \n in Neighborhood 

df_pct['Neighborhood'] = df_pct.Neighborhood.str.replace("\n","")

# indexes reset

df_pct = df_pct.reset_index(drop=True)
df_pct
```

```python
# obtain indices 'Not assigned' for Neighborhood
indexes_neighborhood= df_pct[df_pct['Neighborhood']=='Not assigned'].index
indexes_neighborhood
```

```python
# List Neighborhoods 'Not assgined' with a Borough given
for i in range(212):
    if df_pct.Neighborhood[i]=='Not assigned':
        df_pct.Neighborhood[i] = df_pct.Borough[i]
        
df_pct
```

```python
# Put together the Neighborhood of same Postcodes lines in one line. 
df_pct_fin = df_pct.groupby(['Postcodes','Borough'])['Neighborhood'].apply(','.join).reset_index()
df_pct_fin
```

```python
# Getting venues category  by way of a function 

def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = res['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# columns only
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# only one category per a row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# columns cleaning up
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
## doing the same process for the neighborhoods- a function

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        #  API URL request
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        #  GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # relevant information for every nearby venue to be returned
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# looking for the rest of teh venues in the  Neighbohoods
to_venues = getNearbyVenues(names=merged_df_pct['Neighborhood'],
                                   latitudes=merged_df_pct['Latitude'],
                                   longitudes=merged_df_pct['Longitude']
                                  )
```

```python
#Understanding every  Neighborhood

# applying one hot encoding to transform cat variables
to_onehot = pd.get_dummies(to_venues[['Venue Category']], prefix="", prefix_sep="")

# returning neighborhood column in the dataframe
to_onehot['Neighborhood'] = to_venues['Neighborhood'] 

# get neighborhood col as the first one
column_list = to_onehot.columns.tolist()
column_number = int(column_list.index('Neighborhood'))
column_list = [column_list[column_number]] + column_list[:column_number] + column_list[column_number+1:]
to_onehot = to_onehot[column_list]

to_onehot.head()
```

```python
to_onehot.shape
```

```python
## rows  grouped  by neighborhood and measuring the mean of  occurrence of every category

to_grouped = to_onehot.groupby('Neighborhood').mean().reset_index()
to_grouped
```

```python
to_grouped.shape
```

```python
# list every neighborhood  with their top 3 most common venues

num_top_venues = 3

for hood in to_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = to_grouped[to_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# sorting out the venues in descending order.

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#a new dataframe displaying the top 5 venues for every neighborhood.

num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# columns as per the number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
nbs_venues_sorted = pd.DataFrame(columns=columns)
nbs_venues_sorted['Neighborhood'] = to_grouped['Neighborhood']

for ind in np.arange(to_grouped.shape[0]):
    nbs_venues_sorted.iloc[ind, 1:] = return_most_common_venues(to_grouped.iloc[ind, :], num_top_venues)

nbs_venues_sorted
```

```python
#Clustering  with K-means to create 3 neigborhood clusters
## Run k-means to cluster the neighborhood into 5 clusters.

#  number of clusters
kclusters = 3

to_grouped_clustering = to_grouped.drop('Neighborhood', 1)

# k-means 
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(to_grouped_clustering)

#  cluster labels cretaed for every row in df
kmeans.labels_[0:10] 
```

```python
# labeling clusters
to_merged = merged_df_pct.head(99)
to_merged['Cluster Labels'] = kmeans.labels_

#  to_grouped merged  with to_data so every neighborhood could have latitude and longitude 
to_merged = to_merged.join(nbs_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

to_merged.head()
```

```python
# map creation
map_cls = folium.Map(location=[latitude, longitude], zoom_start=11)

# clusters` color 
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# adding markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(to_merged['Latitude'], to_merged['Longitude'], to_merged['Neighborhood'], to_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_cls)
       
map_cls
```
